## Import Library

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dataset = pd.read_csv("crypto-query-tweets.csv")
```

## Grafik Batang 
Grafik Pie cocok digunakan pada data
categorical. Grafik ini biasa digunakan untuk melihat distribusi.

```python
plt.figure(figsize=(5,5))
plt.title('Diagram Pie')
plt.pie(dataset['verified'].value_counts(), labels=['Verified','Not Verified'],startangle=30, autopct='%.1f%%', explode=(0.3,0))
plt.legend(loc = 'upper left')
plt.show()
```

```python
df= pd.read_csv('covid_19_india.csv')
```

## Scatter Plot
Scatter Plot atau yang memiliki nama lain plot pencar merupakan
diagram grafis yang dibangun dari dua sumbu X dan Y (variabel X dan variabel
Y).
scatter plot digunakan untuk melihat kelompok data dalam
clustering dan melihat outliers.

```python
plt.figure(figsize=(6,10))
plt.title('Scatter Plot')
sns.scatterplot(data = df, x = "Time", y = "ConfirmedIndianNational", label='ConfirmedIndianNational')
plt.legend()
plt.show()
```

## Grafik Batang 
Grafik batang sangat cocok untuk memvisualisasi data kategorikal. Grafik ini biasa digunakan untuk membandingkan kategori.

```python
d= pd.read_csv('covid_vaccine_statewise.csv')
plt.figure(figsize=(7,5))
plt.title('Grafik Batang')
plt.xlabel('State')
plt.ylabel('Jumlah State')
d.State.value_counts().plot.bar()
plt.legend()
plt.show()
```
